Immputing Numerical values using KNN approch and Catergorical Objects using Random imputaion approch

```python
arrC=['AttendingPhysician','OperatingPhysician','ClmDiagnosisCode_2','ClmDiagnosisCode_3','ClmDiagnosisCode_4','ClmDiagnosisCode_5','ClmDiagnosisCode_6','ClmDiagnosisCode_7','ClmDiagnosisCode_8','ClmDiagnosisCode_9']
arrN=['DeductibleAmtPaid','ClmProcedureCode_1']

#imputation

#numerical data
from sklearn.impute import KNNImputer

knn_imputer = KNNImputer(n_neighbors=5)

chunk_size = 1000  #will impute 1000 at a time since dataset is large it takes time


for i in range(0, len(fdata1), chunk_size):
    chunk_data = fdata1.iloc[i:i + chunk_size]
    imputed_chunk = knn_imputer.fit_transform(chunk_data[arrN])
    
    fdata1.loc[i:i + chunk_size - 1, arrN] = imputed_chunk

#categorial data
import random


for i in arrC:
    categories = fdata1[i].dropna().unique()
    fdata1[i].fillna(random.choice(categories), inplace=True)

```
