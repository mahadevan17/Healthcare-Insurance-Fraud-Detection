```python
i=2

while(i<=3):
    in_pca=PCA(n_components=i)

    df_in=in_pca.fit_transform(inX)

    inX_train,inX_test,iny_train,iny_test=train_test_split(df_in,iny,test_size=0.2,random_state=42)

    inX_train,iny_train=SMOTE().fit_resample(inX_train,iny_train) 

    l1=LogisticRegression(random_state=42)

    l1.fit(inX_train,iny_train)

    y_predicted=l1.predict(inX_test)

    accuracy=accuracy_score(iny_test,y_predicted)
    print(i,"th component accuracy for logistic Regression: ",accuracy*100)
    i+=1
```
